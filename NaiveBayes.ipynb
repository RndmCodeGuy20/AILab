```python
import pandas as pd
import numpy as np
```

```python
daywise_df = pd.read_csv('./NaiveBayes.csv')

daywise_df
```

```python
def createOutlookTable():
    sunnyYesCount = daywise_df[['outlook', 'play']].loc[(daywise_df.outlook == 'Sunny') & (daywise_df.play == 'Yes')].shape[0]
    overcastYesCount = daywise_df[['outlook', 'play']].loc[(daywise_df.outlook == 'Overcast') & (daywise_df.play == 'Yes')].shape[0]
    rainYesCount = daywise_df[['outlook', 'play']].loc[(daywise_df.outlook == 'Rain') & (daywise_df.play == 'Yes')].shape[0]

    sunnyNoCount = daywise_df[['outlook', 'play']].loc[(daywise_df.outlook == 'Sunny') & (daywise_df.play == 'No')].shape[0]
    overcastNoCount = daywise_df[['outlook', 'play']].loc[(daywise_df.outlook == 'Overcast') & (daywise_df.play == 'No')].shape[0]
    rainNoCount = daywise_df[['outlook', 'play']].loc[(daywise_df.outlook == 'Rain') & (daywise_df.play == 'No')].shape[0]

    yes = pd.Series([sunnyYesCount, overcastYesCount, rainYesCount])
    no = pd.Series([sunnyNoCount, overcastNoCount, rainNoCount])
    # print(yesSeries, "\n", noSeries)

    outlook_df = pd.DataFrame(index=['Sunny', 'Overcast', 'Rain'])
    outlook_df['yes'] = [sunnyYesCount, overcastYesCount, rainYesCount]
    outlook_df['no'] = [sunnyNoCount, overcastNoCount, rainNoCount]

    outlook_df['p(yes)'] = outlook_df['yes'] / outlook_df['yes'].sum()
    outlook_df['p(no)'] = outlook_df['no'] / outlook_df['no'].sum()
    
    return outlook_df

outlook_df = createOutlookTable()
outlook_df
```

```python
def createTempTable():
    hotYesCount = daywise_df[['temp', 'play']].loc[(daywise_df.temp == 'Hot') & (daywise_df.play == 'Yes')].shape[0]
    mildYesCount = daywise_df[['temp', 'play']].loc[(daywise_df.temp == 'Mild') & (daywise_df.play == 'Yes')].shape[0]
    coolYesCount = daywise_df[['temp', 'play']].loc[(daywise_df.temp == 'Cool') & (daywise_df.play == 'Yes')].shape[0]

    hotNoCount = daywise_df[['temp', 'play']].loc[(daywise_df.temp == 'Hot') & (daywise_df.play == 'No')].shape[0]
    mildNoCount = daywise_df[['temp', 'play']].loc[(daywise_df.temp == 'Mild') & (daywise_df.play == 'No')].shape[0]
    coolNoCount = daywise_df[['temp', 'play']].loc[(daywise_df.temp == 'Cool') & (daywise_df.play == 'No')].shape[0]

    temp_df = pd.DataFrame(index=['Hot', 'Mild', 'Cool'])
    temp_df['yes'] = [hotYesCount, mildYesCount, coolYesCount]
    temp_df['no'] = [hotNoCount, mildNoCount, coolNoCount]

    temp_df['p(yes)'] = temp_df['yes'] / temp_df['yes'].sum()
    temp_df['p(no)'] = temp_df['no'] / temp_df['no'].sum()
    
    return temp_df

temp_df = createTempTable()
temp_df
```

```python
def createHumTable():
    highYesCount = daywise_df[['humidity', 'play']].loc[(daywise_df.humidity == 'High') & (daywise_df.play == 'Yes')].shape[0]
    normalYesCount = daywise_df[['humidity', 'play']].loc[(daywise_df.humidity == 'Normal') & (daywise_df.play == 'Yes')].shape[0]

    highNoCount = daywise_df[['humidity', 'play']].loc[(daywise_df.humidity == 'High') & (daywise_df.play == 'No')].shape[0]
    normalNoCount = daywise_df[['humidity', 'play']].loc[(daywise_df.humidity == 'Normal') & (daywise_df.play == 'No')].shape[0]

    hum_df = pd.DataFrame(index=['High', 'Normal'])
    hum_df['yes'] = [highYesCount, normalYesCount]
    hum_df['no'] = [highNoCount, normalNoCount]

    hum_df['p(yes)'] = hum_df['yes'] / hum_df['yes'].sum()
    hum_df['p(no)'] = hum_df['no'] / hum_df['no'].sum()
    
    return hum_df

hum_df = createHumTable()
hum_df
```

```python
def createWindTable():
    weakYesCount = daywise_df[['wind', 'play']].loc[(daywise_df.wind == 'Weak') & (daywise_df.play == 'Yes')].shape[0]
    strongYesCount = daywise_df[['wind', 'play']].loc[(daywise_df.wind == 'Strong') & (daywise_df.play == 'Yes')].shape[0]

    weakNoCount = daywise_df[['wind', 'play']].loc[(daywise_df.wind == 'Weak') & (daywise_df.play == 'No')].shape[0]
    strongNoCount = daywise_df[['wind', 'play']].loc[(daywise_df.wind == 'Strong') & (daywise_df.play == 'No')].shape[0]

    wind_df = pd.DataFrame(index=['Weak', 'Strong'])
    wind_df['yes'] = [weakYesCount, strongYesCount]
    wind_df['no'] = [weakNoCount, strongNoCount]

    wind_df['p(yes)'] = wind_df['yes'] / wind_df['yes'].sum()
    wind_df['p(no)'] = wind_df['no'] / wind_df['no'].sum()
    
    return wind_df

wind_df = createWindTable()
wind_df
```

```python
def createPlayTable():
    YesCount = daywise_df[['play']].loc[(daywise_df.play == 'Yes')].shape[0]
    NoCount = daywise_df[['play']].loc[(daywise_df.play == 'No')].shape[0]

    play_df = pd.DataFrame(index=['Yes', 'No'])
    
    play_df['count'] = [YesCount, NoCount]
    play_df['p(yes)/p(no)'] = [play_df['count'].loc['Yes'] / play_df['count'].sum(), play_df['count'].loc['No'] / play_df['count'].sum()]

    return play_df

play_df = createPlayTable()
play_df
```

```python
today = ('Sunny', 'Mild', 'High', 'Weak')
today
```

```python
pYes = 1
pNo = 1

for i in today:
    if i in list(outlook_df.index):
        pYes *= outlook_df['p(yes)'].loc[i]
        pNo *= outlook_df['p(no)'].loc[i]
    elif i in list(temp_df.index):
        pYes *= temp_df['p(yes)'].loc[i]
        pNo *= temp_df['p(no)'].loc[i]
    elif i in list(hum_df.index):
        pYes *= hum_df['p(yes)'].loc[i]
        pNo *= hum_df['p(no)'].loc[i]
    elif i in list(wind_df.index):
        pYes *= wind_df['p(yes)'].loc[i]
        pNo *= wind_df['p(no)'].loc[i]

# print(pYes, pNo)

pYes *= play_df['p(yes)/p(no)'].loc['Yes']
pNo *= play_df['p(yes)/p(no)'].loc['No']

pYesFin = (pYes / (pYes + pNo))
pNoFin = (pNo / (pYes + pNo))

# print(pYes, pNo)

probPlay = np.array([pYesFin, pNoFin])
print(probPlay)

if probPlay.argmax() == 1:
    print("Player will probably not play")
else:
    print("Player will probably play")
```
